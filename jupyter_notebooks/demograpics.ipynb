```python
def renameColumns(df):
    renamed_df = df.rename(columns={
    "zip code tabulation area": "zipcode",
    "NAME": "name", 
    "B01003_001E": "population",
    "B01002_001E": "median_age",
    "B06011_001E": "median_income",
    "B19013_001E": "household_income",
    "B19301_001E": "per_capita_income",
    "B25104_001E": "monthly_housing_cost",
    "B25010_001E": "avg_household_size_by_tenure"
                                     })
    return renamed_df
```

```python
census_df_2017 = renameColumns(census_df_2017)
census_df_2016 = renameColumns(census_df_2016)
census_df_2015 = renameColumns(census_df_2015)
census_df_2014 = renameColumns(census_df_2014)
census_df_2013 = renameColumns(census_df_2013)
census_df_2012 = renameColumns(census_df_2012)
```

```python
zip_df = pd.read_csv("../resources/free-zipcode-database-Primary.csv")
```

```python
zip_df['Zipcode'] = zip_df['Zipcode'].map(lambda x: "0"*(5-len(str(x)))+str(x))
```

```python
zip_df = zip_df.rename(columns={
    "Zipcode": "zipcode",
    "City": "city",
    "State": "state_abbr",
    "ZipCodeType": "zipcode_type",
    "Lat": "lat",
    "Long": "lon"
})
```

```python
zip_df = zip_df[["zipcode",
                 "city",
                 "state_abbr",
                 "zipcode_type",
                 "lat",
                 "lon"
                ]]
```

```python
states_df = pd.read_csv("../resources/states.csv")
states_df = states_df.rename(columns={
    "State": "state",
    "Abbreviation": "state_abbr"
})
states_df.head()
```

```python
join_df = zip_df.join(states_df.set_index("state_abbr"), on="state_abbr")
```

```python
demo_2017_df = join_df.join(census_df_2017.set_index("zipcode"),on="zipcode" )
demo_2016_df = join_df.join(census_df_2016.set_index("zipcode"),on="zipcode" )
demo_2015_df = join_df.join(census_df_2015.set_index("zipcode"),on="zipcode" )
demo_2014_df = join_df.join(census_df_2014.set_index("zipcode"),on="zipcode" )
demo_2013_df = join_df.join(census_df_2013.set_index("zipcode"),on="zipcode" )
demo_2012_df = join_df.join(census_df_2012.set_index("zipcode"),on="zipcode" )
```

```python
demo_2017_df['city_state'] = demo_2017_df['city']+', ' + demo_2017_df['state_abbr']
demo_2016_df['city_state'] = demo_2016_df['city']+', ' + demo_2016_df['state_abbr']
demo_2015_df['city_state'] = demo_2015_df['city']+', ' + demo_2015_df['state_abbr']
demo_2014_df['city_state'] = demo_2014_df['city']+', ' + demo_2014_df['state_abbr']
demo_2013_df['city_state'] = demo_2013_df['city']+', ' + demo_2013_df['state_abbr']
demo_2012_df['city_state'] = demo_2012_df['city']+', ' + demo_2012_df['state_abbr']
```

```python
def filter_df(df):
    new_df = df.loc[df["zipcode_type"]=="STANDARD"].loc[df["population"]>0]
    return new_df
```

```python
demo_2017_df = filter_df(demo_2017_df)
demo_2016_df = filter_df(demo_2016_df)
demo_2015_df = filter_df(demo_2015_df)
demo_2014_df = filter_df(demo_2014_df)
demo_2013_df = filter_df(demo_2013_df)
demo_2012_df = filter_df(demo_2012_df)
```
