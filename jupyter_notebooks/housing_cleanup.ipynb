```python
def renameColumn(name):
    name = name.lower()
    if name[-1] == " ":
        name = name[:-1]
    name = name.replace(" ", "_")
    return name
```

```python
new_zip_df.columns = [renameColumn(col_name) for col_name in new_zip_df.columns]
new_zip_df = new_zip_df[:-1]
new_zip_df = new_zip_df.rename(columns={'month':'date'})
# new_zip_df['date'] = new_zip_df['date'].apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
# new_zip_df['year_month'] = new_zip_df['date'].apply(lambda x: f"{x.year}/{x.month}")
```

```python
new_zip_df['date'] = new_zip_df['date'].apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
new_zip_df['year_month'] = new_zip_df['date'].apply(lambda x: f"{x.year}/{x.month}")
```

```python
# new_zip_df["month"] = new_zip_df["date"].apply(lambda x: f"{x.month}")
# new_zip_df["year"] = new_zip_df["date"].apply(lambda x: f"{x.year}")
```

```python
new_zip_df["city"] = new_zip_df["zipname"].apply(lambda x: x.split(",")[0])
new_zip_df["state"] = new_zip_df["zipname"].apply(lambda x: x.split(",")[1][1:])
```

```python
new_zip_df.columns
```

```python
new_zip_df = new_zip_df[["year_month",
    "zipcode",
    "city",
    "state",
    "median_listing_price",
    "avg_listing_price",
    "price_increase_count",
    "price_decrease_count",             
    "new_listing_count",
    "total_listing_count",
    "days_on_market",
    ]]
```

```python
def formatZip(zipcode):
    zipcode = str(int(zipcode))
    if len(zipcode) == 4:
        zipcode = "0"+zipcode
    return zipcode
```

```python
new_zip_df.zipcode = new_zip_df.zipcode.map(lambda x: formatZip(x))
```
